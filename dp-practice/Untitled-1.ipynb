```typescript
const prices = [7, 6, 4, 3, 1];
const getPrices = (prices: Array<number>) => {
  if (prices.length === 0) return 0;
  let sum = 0;
  // 当前持有的股票
  let cur = undefined;
  for (let i = 0; i < prices.length - 1; i++) {
    // console.log(cur);
    if (prices[i] < prices[i + 1]) {
      if (cur === undefined) {
        cur = i;
      }
      continue;
    } else if (cur !== undefined) {
      sum += prices[i] - prices[cur];
      cur = undefined;
    }
  }
  // console.log(cur, prices[prices.length - 2], prices[prices.length - 1], sum);
  if (
    cur !== undefined &&
    prices[prices.length - 2] < prices[prices.length - 1]
  ) {
    sum += prices[prices.length - 1] - prices[cur];
  }
  return sum;
};

console.log(getPrices(prices));
```

```typescript
const prices = [1, 2, 4];
const getMaxValue = (prices: Array<number>) => {
  const dp = Array.from({ length: prices.length }, () => [0, 0]);
  if(prices.length <= 1) return 0;
  if(prices.length <= 2){
    return Math.max(0, prices[1] - prices[0])
  }
  dp[0][0] = 0;
  dp[0][1] = -prices[0];
  dp[1][0] = Math.max(0, prices[1] - prices[0])
  dp[1][1] = Math.max(-prices[1], -prices[0])
  for (let i = 2; i < prices.length; i++) {
    // 未持有
    dp[i][0] = Math.max(dp[i - 1][0], dp[i - 1][1] + prices[i]);
    // 持有
    dp[i][1] = Math.max(dp[i - 1][1], dp[i - 2][0] - prices[i]);
  }
  console.log(dp, Math.max(...dp[prices.length - 1]));
  return Math.max(...dp[prices.length - 1]);
};

console.log(getMaxValue(prices));
```

```typescript
const obstacleGrid = [
  [0, 1],
  [0, 0],
];
const getWays = (input: Array<Array<number>>) => {
  const dp = Array.from({ length: input.length }, () =>
    Array.from({ length: input[0].length }, () => 0),
  );
  // init
  dp[0][0] = input[0][0] === 1 ? 0 : 1;
  for (let i = 1; i < input.length; i++) {
    if (input[i][0]) {
      dp[i][0] = 0;
      continue;
    }
    dp[i][0] = dp[i - 1][0];
  }
  for (let j = 1; j < input[0].length; j++) {
    if (input[0][j]) {
      dp[0][j] = 0;
      continue;
    }
    dp[0][j] = dp[0][j - 1];
  }
  console.log(dp);
  for (let i = 1; i < input.length; i++) {
    for (let j = 1; j < input[0].length; j++) {
      if (input[i][j]) {
        dp[i][j] = 0;
        continue;
      }
      dp[i][j] = dp[i - 1][j] + dp[i][j - 1];
    }
  }
  // console.log(dp);
  return dp[input.length - 1][input[0].length - 1];
};

console.log(getWays(obstacleGrid));
```
